```python
import io
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://www.conab.gov.br/info-agro/safras/serie-historica-das-safras/item/download/34775_4310db971a5045e5dc486071e4d742ba'
s=requests.get(url).content
xls=pd.ExcelFile(s)
```

```python
# Valores em mil hectares
df = xls.parse(skiprows=5)
df = df.set_index('REGIÃO/UF')
df.dropna(inplace=True)
df = df.transpose()
```
