# SÓ MILHO E TRIGO

```python
import io
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://portaldeinformacoes.conab.gov.br/downloads/arquivos/OfertaDemanda.txt'
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')), sep=';')
```

```python
# remove espaços em branco
df = df.applymap(lambda x: x.strip() if type(x) == str else x)

# Converte sistema numérico
def conversor(df, colunas:list):
    for coluna in colunas:
        try:
            df[coluna] = df[coluna].map(lambda x: float(x.replace(".","").replace(",",".")))
        except:
            pass
    return df

df = conversor(df, df.columns)
df
```
