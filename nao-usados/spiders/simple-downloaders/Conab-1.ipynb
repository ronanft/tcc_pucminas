```python
import io
import requests
import pandas as pd
import numpy as np
```

```python
url = 'https://portaldeinformacoes.conab.gov.br/downloads/arquivos/LevantamentoGraos.txt'
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')), sep=';')
```

```python
df=df.set_index('ano_agricola')
```

```python
# remove espaços em branco
df = df.applymap(lambda x: x.strip() if type(x) == str else x)
```

```python
df.produto.value_counts()
```

```python
df_agg = df.groupby(['produto','ano_agricola','id_levantamento']).agg({'area_plantada_mil_ha':'sum','producao_mil_t':'sum','produtividade_mil_ha_mil_t':'mean'})
```
