#### Todavia, o ideal é indexar o preço à inflação e/ou ao dólar

```python
# Usaremos primeiro a indexação ao IGP-M
# Dados exportados no formato .csv em 03.01.2021 do endereço http://www.ipeadata.gov.br/Default.aspx

igpm = pd.read_csv('../dados/ipeadata[03-01-2021-06-33].csv', dtype={'Data':str})
```

```python
igpm.columns = ['data', 'igpm', 'nan']
igpm.drop(columns=['nan'], axis=1, inplace=True)
```

```python
# Convert coluna data para datetime e seta como índice
def conv(t):
    return datetime.strptime(str(t),'%Y.%m')

igpm.index = igpm.data.apply(conv)
igpm.drop(columns=['data'], inplace=True, axis=1)
igpm = igpm.resample('MS').first()
```

```python
# Adiciona dados do igpm à tabela de cotações
df_dados = df_dados.join(igpm, how="inner")
```

```python
# Cria coluna indexadora
df_dados['igpm_idx'] = (df_dados.igpm/100)+1
df_dados['igpm_cum'] = df_dados.igpm_idx[::-1].cumprod()

# # Calcula o preço nacional indexado
df_dados['preco_atual'] = df_dados.nacional * df_dados.igpm_cum
```

```python
df_dados.drop(columns=['igpm','igpm_idx','igpm_cum'], axis=1, inplace=True)
```

```python
def adf_test(X):
    result = adfuller(X)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
```

```python
# ADF TEST - Série de Preço indexado
adf_test(df_dados.preco_atual)
```

```python
def acf_pacf(series):
    fig = pyplot.figure(figsize=(6,8))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(series,lags=10, ax=ax1)
    fig.tight_layout()

    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(series,lags=10, ax=ax2)
    fig.tight_layout()
    pyplot.show()
    
    # Scatter plot entre (t) e (t-1)
    pyplot.figure(figsize=(6,6))
    pd.plotting.lag_plot(series, lag=1)
```

```python
# ACF e PACF - preço dolarizado
acf_pacf(df_dados.preco_atual)
```

```python
# Análise dos dados nacionais COM indexação ao IGPM
series = df_dados.preco_atual

# Decomposição da série em tendência, sazonalidade e resíduos
rcParams['figure.figsize'] = 11, 9
result = seasonal_decompose(series, model='additive')
pyplot.figure()
result.plot()
pyplot.show()
```

```python
a_r = result.resid.max()-result.resid.min()
a_s = result.seasonal.max()-result.seasonal.min()
print(f'Amplitude residual de feijão: {"{:.2f}".format(a_r)}')
print(f'Amplitude sazonal de feijão: {"{:.2f}".format(a_s)}')
print(f'Relação amplitude residual / sazonal:  {"{:.2f}".format(a_r/a_s)}')
```

```python
# Mês mais "caro" é junho
result.seasonal.sort_values(ascending=False)[:16]
```

```python
# Apesar da decomposição automática apontar novembro como o mês mais barato (o que está correto conforme a indexação),
# a média de janeiro é inferior a novembro em termos nominais.

# Mês mais barato
barato = df_dados.nacional.groupby(df_dados.nacional.index.month).agg('mean').idxmin()

# Mês mais caro
caro = df_dados.nacional.groupby(df_dados.nacional.index.month).agg('mean').idxmax()

# Retorna o número com apenas 2 casas decimais (função meramente estética)
def two_dec(f):
    return (int(f*10000))/100

# Proporcao
proporcao = two_dec(1 - df_dados.nacional.groupby(df_dados.nacional.index.month).agg('mean').min()/df_dados.nacional.groupby(df_dados.nacional.index.month).agg('mean').max())
```

```python
print(f'O mês mais barato é o mês {barato}')
print(f'O mês mais caro é o mês {caro}')
print(f'Na proporção de {proporcao}%')
```

```python
# Buscando tendência relevante entre os anos
pyplot.figure(figsize=(10,4))
df_dados.preco_atual.groupby(df_dados.preco_atual.index.year).agg('mean').plot()

# A tendência de alta é expurgada quando se retira a inflação IGPM da série.
# Há, visualmente, uma alternância entre anos de preço alto e baixo. Talvez porque os produtores se vêem
# impelidos a plantar no ano seguinte quando o preço de determinado ano está alto e dissuadidos quando o preço está baixo.
```

```python
# Coeficiente de variação da série agregada anualizada
cv = two_dec(df_dados.preco_atual.groupby(df_dados.preco_atual.index.year).agg('mean').std() / df_dados.preco_atual.groupby(df_dados.preco_atual.index.year).agg('mean').mean())
print(f'Coeficiente de variação de {cv}%')
```

### Simulação de estratégia de comercialização por intermediário da cadeia produtiva

Trata-se de estratégia teórica, sem possibilidade prática, uma vez que o feijão não pode ser estocado por longos períodos (sem refrigeração e controle de umidade), pois é o mais perecível dos grãos, devido ao caruncho.

Assim, comprar o grão para vender 6 meses depois é possível na teoria, mas não na prática.
Todavia, para fins acadêmicos, realizaremos a simulação.

```python
# Simulação de compra no mês de janeiro para venda no mês de junho
# Foi utilizada a série desindexada, nominal, uma vez que se trata de estratégia de curto prazo, dentro do próprio ano

ganho_est = []

mes_venda = 6
mes_compra = 1

for year in df_dados.nacional.index.year.unique():
    cond1 = df_dados.nacional.index.year == year
    cond2 = df_dados.nacional.index.month == mes_venda
    cond3 = df_dados.nacional.index.month == mes_compra
    ganho_est.append(two_dec(df_dados.nacional[cond2 & cond1][0] / df_dados.nacional[cond3 & cond1][0]))
```

```python
# Resultado do backtest de ganho anual para a estratégia acima (fração * 100)
ganho_est
```

```python
# Demais estatísticas de ganho
npganho = np.array(ganho_est)
npganho = npganho / 100

# Ganho acumulado no período de 16 anos
print('Ganho do capital em todo o período: (1 = 100%)')
cum_ganho = npganho.cumprod()[-1]
cum_ganho
```

```python
# CAGR
cagr = cum_ganho ** (1/len(npganho))

print('Fator de ganho médio composto anual (1 = 100% de ganho)')
((cagr - 1) * 100) - 1
```

```python
# % de ganhos
a = lambda x: 1 if x > 100 else 0

print('Percentual de anos com ganho')
sum(list(map(a, ganho_est)))/len(ganho_est)*100
```

### Análise

* Há uma diferença de preços relevante entre o mês mais barato e o mais caro (janeiro e junho - em termos nominais).

* Da decomposição da série indexada, com a biblioteca Statsmodels, depreende-se que não há tendência no preço.

* Anos caros e baratos se alternam, em uma sazonalidade "anual", possivelmente porque os produtores se vêm estimulados a plantar feijão após um ano de bons preços e desestimulados após um ano de preços baixos).

* Não foi considerado aqui o efeito do preço do feijão na própria inflação que indexou seu preço (esse fator existe e leva a uma retroalimentação da variável). Enfim, em certa medida, a inflação aumenta quando o preço do feijão aumenta.

* O coeficiente de variação dos preços anuais é 46,3% em termos nominais e 31% na série indexada.
